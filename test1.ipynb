```python
def sort_array (arr):
    smallest = arr[0]
    smallest_index=0
    for i in range(1, len(arr)):
        if arr[i]<smallest:
            smallest = arr[i]
            smallest_index = i
    return smallest_index


def sel_sort(arr):
    newArr = []
    for i in range(len(arr)):
        smallest= sort_array(arr)
        newArr.append(arr.pop(smallest))
    return newArr


list = [5,3,2,5,7,9,4,2,1]


print(sort_array(list))
print(sel_sort(list))
```

```python
def last_digit(n1, n2):
    if n2 ==0:
        return 1

    result = n1 ** n2
    result = str(result)
    lst=[]
    for i in result:
        lst.append(i)

    return lst[-1]


print(last_digit(999,0))
```

```python
def print_numbers(n):
    if n > -8:
        print_numbers(n - 1)  # Recursive call
        print(n)

# Example usage
print_numbers(5)
```

```python
def factorial(n):
    if n==0 or n==1:
        return 1
    else:
        return n* factorial(n-1)
    

factorial(5)
```

```python
def smd(n):
    if n<10:
        return n
    else:
        return n % 10 + smd(n // 10)
    
print(smd(456))
```

```python
def sum_recursive(arr, index=0):
    # Base case: if the index is equal to the length of the array, return 0
    if index == len(arr):
        return 0
    else:
        # Recursive case: sum the current element and the sum of the rest of the array
        return arr[index] + sum_recursive(arr, index + 1)

# Example usage
array = [1, 2, 3, 4, 5,99]
result = sum_recursive(array)
print(f"The sum of the array is: {result}")
```

```python
def quick_sort(arr):
    sml = []
    big = []
    pivot = arr[0]
    if len(arr)==1:
        return arr
    else:
        for i in range (1, len(arr)):
            if arr[i]<pivot:
                sml.append(i)
            elif arr[i]>pivot:
                big.append(i)
    return quick_sort(sml) + pivot + quick_sort(big)
        
```
